```python
# tentativo di generare dati medi (punti) con barra di errore
import altair as alt
import pandas as pd
import numpy as np

# generate some data points with uncertainties
np.random.seed(0)
x = createList(0,568)
yerr = 0.2

# set up data frame
source = pd.DataFrame({"x":x, "y":m_rad, "yerr":yerr})

# the base chart
base = alt.Chart(source).transform_calculate(
    ymin="datum.y-datum.yerr",
    ymax="datum.y+datum.yerr"
)

# generate the points
points = base.mark_point(
    filled=True,
    size=50,
    color='black'
).encode(
    x=alt.X('x', scale=alt.Scale(domain=(0, 568))),
    y=alt.Y('y', scale=alt.Scale(domain=(min(m_rad), max(m_rad))))
)

# generate the error bars
errorbars = base.mark_errorbar().encode(
    x="x",
    y="ymin:Q",
    y2="ymax:Q"
)

# commento il grafico (illeggibile)
#points + errorbars
```
