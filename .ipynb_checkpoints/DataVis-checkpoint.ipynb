```python
#Seaborn Ã¨ una libreria che dipende da pandas, numpy e matplotilib, per usarla vanno importati tutti
import seaborn as sns
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
#LONG-FORM DATA

flights.head() 

# Si nota la differenza tra ogni mese di un anno
#Colonne=variabili
#righe=osservazioni 
```

```python
#ESEMPIO FLIGHT DATASET LONG-FORM DATA
flights = sns.load_dataset("flights")
sns.relplot(data=flights, x="year", y="passengers", hue="month", kind="line") 
#nei long-form dataset le colonne devono essere esplicitamente assegnate alle variabili degli assi
```

```python
#VISUALIZZAZIONE DEI WIDE-FORM DATA, righe contengono il nÂ° di passeggeri per ogni mese, righe diverse=anni diversi
flights_wide.head()
#si nota la differenza per ogni anno
```

```python
# WIDE-FORM DATA
flights_wide = flights.pivot(index="year", columns="month", values="passengers")
sns.relplot(data=flights_wide, kind="line")
```

```python
# DATASET IN FORMA DISORDINATA
anagrams = sns.load_dataset("anagrams")
anagrams.head()
```

```python
# Va trasformato in un long o wide form dataset 
# si usa il metodo pandas.DataFrame.melt()
anagrams_long = anagrams.melt(id_vars=["subidr", "attnr"], var_name="solutions", value_name="score")
anagrams_long.head()
```

```python
# Con Pandas ci sono degli operatori di raggruppamento e divisione per i dati 
flights_avg = flights.groupby("year").mean() # Groupby()
sns.relplot(data=flights_avg, x="year", y="passengers", kind="line")
#questo grafico consente di vedere il numero di passeggeri annuali e come questo trend cambia
```

```python
#itâs possible to pass vectors of data directly as arguments to x, y, and other plotting variables. 
#If these vectors are pandas objects, the name attribute will be used to label the plot:
year = flights_avg.index
passengers = flights_avg["passengers"]
sns.relplot(x=year, y=passengers, kind="line")
# il risultato Ã¨ identico ma Ã¨ cambiato il modo di creare il grafico stesso, vengono passati gli attributi alle variabili
```

ESEMPI

```python
#ESEMPIO PENGUINS DATASET
#HISTPLOT 
penguins = sns.load_dataset("penguins") #import del dataset penguins da seaborn
sns.histplot(data=penguins, x="flipper_length_mm", hue="species", multiple="stack")
```

```python
#KDEPLOT
sns.kdeplot(data=penguins, x="flipper_length_mm", hue="species", multiple="stack")
```

```python
#SCATTERPLOT, HISTPLOT
f, axs = plt.subplots(1, 2, figsize=(8, 4), gridspec_kw=dict(width_ratios=[4, 3]))
sns.scatterplot(data=penguins, x="flipper_length_mm", y="bill_length_mm", hue="species", ax=axs[0])
sns.histplot(data=penguins, x="species", hue="species", shrink=.8, alpha=.8,legend=False, ax=axs[1])
f.tight_layout()
```
